```
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```
from google.colab import drive
drive.mount("/drive")
x_train = pd.read_csv('/drive/My Drive/Colab Notebooks/proj1_data/x_train.csv')
x_test = pd.read_csv('/drive/My Drive/Colab Notebooks/proj1_data/x_test.csv')
y_train = pd.read_csv('/drive/My Drive/Colab Notebooks/proj1_data/y_train.csv')

x_train.head()

x_test.head()

y_train.head()
```

```
x_test
```

```
def LinearRegressionModel(X, y):
    # Inputs:
    regr = LinearRegression()
    clf = regr.fit(X,y)   
    # IMPLEMENT THIS METHOD USING sklearn.linear_model.LinearRegression 
    return clf
```

```
y_train_new = y_train.drop(columns=['Sl. No.'])
y_train_new #train with this

y_droppedcol = y_train.drop(columns=['Output'])
y_droppedcol #use this later
```

```
first_model = LinearRegressionModel(x_train, y_train_new)
first_model_predict = first_model.predict(x_train)

first_model_predict #attach this to sl column
```

```
#checking_preds = pd.DataFrame({'Actual': y_train_new, 'Predicted': first_model_predict})
first_predictions = pd.DataFrame(data=first_model_predict, columns=["predictions"])
first_predictions

```

```
y_train_new
```

```
print('Mean Squared Error of Training:', metrics.mean_squared_error(y_train_new, first_model_predict))
```
